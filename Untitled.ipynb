#### Column Description
+ InvoiceNo: Unique identifier for each transaction.
+ StockCode: Unique product code identifying each item.
+ Description: Name of the product.
+ Quantity: Number of units purchased in a transaction.
+ InvoiceDate: DateTime	Timestamp when the transaction occurred. Useful for trend and seasonal analysis.
+ UnitPrice: Price per unit of the product.
+ CustomerID: Unique identifier for each customer.
+ Country: The country where the transaction occurred.

```python
Q1 = df[['Quantity', 'UnitPrice']].quantile(0.25)
Q3 = df[['Quantity', 'UnitPrice']].quantile(0.75)
```

```python
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
df = df[~((df[['Quantity', 'UnitPrice']] < (Q1 - 1.5 * IQR)) | (df[['Quantity', 'UnitPrice']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```
